### 나만의 function을 만드는 방법 
- 키워드를 넣는다 - def (definding function)
- 함수에 이름을 붙여 준다. - def "function name"
- 함수 이름에 뒤에 입력값을 위한 괄호를 넣어준다.
- 함수의 정의를 마무리 짓기 위해 ":"(콜론)을 넣어준다. - 해당 줄 다음에 오거나 들여쓰기 되는 모든 것은 함수에 속한다라는 의미 

```python
# structure 
def my_fucntion():
    # Do this
    # Then do this
    # Finally do this
```
