4. 게임이 종료되면 최종적으로 맞춘 점수 보여주기, 그리고 게임 다시시작 여부 묻기 

### 우리가 해야할 final project Higer-Lower Game 방법 
- 게임 목적 - 누구의 인스타 팔로워가 더 많을까 맞추기 
1. 로고 출력 
2. 비교 상대 랜덤으로 출력 
    - Compare A: name, description, country vs Against B: name, description, country (vs는 로고 사용) 
    - who has more followers? Type 'A' or 'B': 입력 문구 출력
3. 만약 팔로워수가 많은 사람을 골랐다면 
    - You're right! Current score: 가 1씩 증가 
    - 만약 B를 맞췄다면가 B를 A로 올림
    - 그리고 다시 랜덤으로 비교 상대 출력 
4. 만약 팔로워수가 적은 사람을 골랐다면
    - Sorry, that's wrong. Final score: 출력 후 게임 종료 

### 코드 작성시 우리가 해야할 사항 
1. 문제를 작게 나누기 - 작은 문제 하나씩 해결하면서 작업  
2. to-do list 작성 
3. 2번 to-do를 바탕으로 주석으로 입력후 코드를 나눠서 입력 
4. 코드를 작성하고 -> 코드를 실행시키고 -> 에러 코드 고치는 작업 반복

```python
import random
from higher_lower import art
from higher_lower import game_data

# 1. 로고 출력 
print(art.logo)              # 시작 로고 

# 2. 팔로우 데이터 불러오기 
data = game_data.data

game_over = False
score = 0
while not game_over:
    # 3. 초반에 랜덤하게 두 대결 상대를 출력 
    compare = random.choice(data)
    print("Compare A: " + compare['name'] + ", " + compare['description'] + ", " + "from " + compare['country'])
    print(art.vs)                # vs 로고 
    against = random.choice(data)
    print("Against B: " + against['name'] + ", " + against['description'] + ", " + "from " + against['country'])
    
    # 4. 선택할 수 있게 input 할 사항 넣기
    choice = input("Who has more followers? Type 'A' or 'B': ")
    if choice == 'A':
        choice_result = compare['follower_count']
        print(choice_result)
    else:
        choice_result = against['follower_count']
        print(choice_result)
        
    if choice_result > compare['follower_count']:
        score += 1
        print(f"you got it Current score: {score}")
    elif choice_result > against['follower_count']:
        score += 1
        print(f"you got it Current score: {score}")
    else:
        print(f"Sorry, that's wrong. Final score:{score}")
        game_over = True
```
